```python
import pandas as pd
```

```python
df = pd.read_csv("data/powerstations.csv")
df['size_mw'] = df['size_kw']/1000
df.head()
```

```python
per_state = df.groupby("state").size_mw.agg("sum")
per_state
per_state.to_csv("data/per_state_capacity.csv")
```

```python
fuel_mix = pd.read_csv("data/national_fuel_mix_12mo.csv")
fuel_mix.head()
```

```python
fuel_mix['Fuel Type'].value_counts()
```

```python
fuel_mix['State'].value_counts()
```

```python
df2 = fuel_mix.groupby(['State', 'Fuel Type']).Supply.sum().groupby(level=0).apply(lambda x: 100 * x / float(x.sum())).reset_index()
df2.columns = ["source", "target", "value"]
df2.to_csv("data/sankey_data.csv")
```

```python
per_country = pd.read_html("https://en.wikipedia.org/wiki/Solar_power_by_country")[1]
print(per_country.columns)
per_country.head()
```

```python
per_country.drop(columns=per_country.columns[-1], axis=1, inplace=True)
```

```python
df = pd.read_csv("data/per-capita-solar.csv")
filt = (df.Entity == "Australia") | (df.Entity == "United States") | (df.Entity == "Spain") | (df.Entity == "India") | (df.Entity == "Portugal") | (df.Entity == "New Zealand") | (df.Entity == "China") | (df.Entity == "Japan")
df2 = df[filt]
df2.columns = ["country", "code", "year", "value"]
df2.head()
```

```python
df2.to_csv("data/per_capita_cleaned.csv", index = False)
```
